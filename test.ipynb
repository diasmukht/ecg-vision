```python
import numpy as np  
import pandas as pd 
```

```python

arr = np.arange(1, 21) 
print("Исходный массив:", arr)


matrix = arr.reshape(4, 5)
print("\nМатрица 4x5:\n", matrix)

row_sums = matrix.sum(axis=1)
print("\nСумма каждой строки:", row_sums)

max_val = matrix.max()
min_val = matrix.min()
print(f"\nМаксимум: {max_val}, Минимум: {min_val}")
```

```python

data = {
    'Name': ['Алихан', 'Мария', 'Болат', 'Саша', 'Айгерим'],
    'Age': [20, 21, 19, 22, 20],
    'GPA': [3.5, 2.8, 3.9, 3.0, 3.2]
}


df = pd.DataFrame(data)
print("Таблица студентов:\n", df)


good_students = df[df['GPA'] > 3.0]
print("\nСтуденты с GPA > 3.0:\n", good_students)


mean_gpa = df['GPA'].mean()
print(f"\nСредний GPA: {mean_gpa}")


count = len(df)
print(f"Всего студентов: {count}")
```

```python
random_data = np.random.randint(10, 100, size=(10, 3))

df_project = pd.DataFrame(random_data, columns=['A', 'B', 'C'])
print("Сгенерированные данные:\n", df_project.head())


stats = df_project.agg(['min', 'max', 'mean'])
print("\nСтатистика:\n", stats)


df_project.to_csv('my_project_data.csv', index=False)
print("\nФайл 'my_project_data.csv' успешно сохранен.")
```
